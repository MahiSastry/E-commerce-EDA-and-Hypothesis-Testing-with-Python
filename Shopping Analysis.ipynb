Experimenting with log transformation

## Summarize Page Views, Durations, and Bounce/Exit Rates for Each Page Category

```python
page_cols = ['Administrative', 'Administrative_Duration', 'Informational',
       'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration']
df[page_cols].describe()
```

```python
page_categories = {
    'Administrative': ['Administrative', 'Administrative_Duration'],
    'Informational': ['Informational', 'Informational_Duration'],
    'ProductRelated': ['ProductRelated', 'ProductRelated_Duration']
}
```

```python
summary_df = pd.DataFrame()
summary_list = []
for category, columns in page_categories.items():
    category_df = df[columns + ['BounceRates', 'ExitRates', 'PageValues']].copy()
    
    
    category_means = category_df.mean()

    category_means['Category'] = category 
    summary_list.append(category_means)

summary_df = pd.concat(summary_list, axis=1).T
summary_df = summary_df[['Category', 'BounceRates', 'ExitRates', 'PageValues']]


print(summary_df)
```

## Visited All Categories

```python
df.columns
```

```python
def check(admin,info,product):
    if admin != 0 and info != 0 and product != 0:
        return 1
    else:
        return 0
```

```python
df['All_Three_Pages'] = df.apply(lambda x : check(x.Administrative, x.Informational, x.ProductRelated), axis = 1)
```

```python
allpages = df.groupby(['All_Three_Pages'])['Revenue'].count().reset_index(name='Count')
allpages.plot(kind = 'bar', color = 'blue')
plt.show()
```

## User Session Lengths and Conversion Rates

```python
df['Total_Session_Duration'] = df['Administrative_Duration'] + df['Informational_Duration'] + df['ProductRelated_Duration']
```

```python
df['Session_Duration_Bins'] = pd.qcut(df['Total_Session_Duration'], q=4, labels=['Short', 'Medium', 'Long', 'Very Long'])
conversion_rates = df.groupby('Session_Duration_Bins')['Revenue'].mean()
```

```python
conversion_rates.plot(kind='bar', title='Conversion Rates by Session Duration')
plt.show()
```

## VisitorType, OperatingSystems, and Region

## Segment Users by TrafficType

```python
traffic_conversion = df.groupby('TrafficType')[['Revenue','Total_Session_Duration']].mean().reset_index().sort_values(by = 'Revenue', ascending = False)
traffic_conversion
```

```python
traffic_grouped = df.groupby('TrafficType').agg({
    'Revenue': 'mean',
    'Total_Session_Duration': 'mean',
    'PageValues': 'mean',
    'Administrative': 'mean',
    'Informational': 'mean',
    'ProductRelated': 'mean'
}).reset_index()

traffic_grouped
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='TrafficType', y='Revenue', data=traffic_grouped, palette='magma')
plt.title('Conversion Rates by Traffic Type')
plt.ylabel('Conversion Rate')
plt.xlabel('Traffic Type')
plt.ylim(0, 1)
plt.show()
```

```python

plt.figure(figsize=(10,6))
sns.barplot(x='TrafficType', y='Total_Session_Duration', data=traffic_grouped, palette='cividis')
plt.title('Average Session Duration by Traffic Type')
plt.ylabel('Average Session Duration (seconds)')
plt.xlabel('Traffic Type')
plt.show()


plt.figure(figsize=(10,6))
sns.barplot(x='TrafficType', y='PageValues', data=traffic_grouped, palette='plasma')
plt.title('Average Page Values by Traffic Type')
plt.ylabel('Average Page Values')
plt.xlabel('Traffic Type')
plt.show()
```

## Insights

1. The dataset consists of 12330 records and 18 columns, no null values
2. May had the highest number of pageviews followed by October.
3. OS 2 was the most used operating system by a wide margin. OS 1 and 3 are close in terms of numbers.
4. Browser 2 was the most used, by a large margin.
5. Region 1 saw the most number of traffic.
6. 3913 visits were made by traffic type 2
7. Returning visitors are the highest in number at 10551.
8. Revenue column has a data imbalance, with most traffic not resulting in conversion(False :10422 , True : 1908)
9. Visual analysis of columns like ('Administrative', 'Administrative_Duration', 'Informational', 
   'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration') showed a heavily right skewed distribution, with a high frequency
   at 0, indicating not a lot of visits or time spent was observed for any of the pages.
10. Informational column has a strong positive relation with All_Three_Pages(binary column to check if the user visited all 3 pages).
11. Revenue and Page value have a moderate positive correlation of 0.4.
12. Exit and Bounce rates have a very high correlation of 0.9. 
13. The most amount of time spent  and number of visits on average is in product related page.
14. 89% of the revenue is generated (0:before or after a Special Day)
15. 'Other' category generates the highest PageValue when it comes to the type of visitor that lands on the page.
16. It is observed that users that spend a lot of time on the website has led to the most number of conversions.
17. A relationship between new visitors and operating system 6 & 8 (correlation = 0.5, 0.47) is observed. Suggesting a preference.
18. Traffic type 16 seems to be generating the most amount of revenue with a duration of 899(assuming it to be in minutes)
19. Traffic type 14 has the highest session duration while type 20 drives the highest page value. 

## Conclusion

The website seems to have a core group of returning visitors who spend limited time on the majority of 
pages but are highly engaged with product-related content. 
Conversion rates are low despite the high traffic, indicating potential areas for optimization in user experience and conversion tactics. 
Understanding visitor behaviors, particularly around specific traffic types, operating systems, and special days, 
could lead to more effective targeting and higher conversion rates.

There is a need to focus on converting the substantial number of visitors who do not currently result in sales, 
possibly through improved content engagement, personalization, and targeted marketing strategies. 
Additionally, the insights around specific traffic types, visitor categories, and operating systems 
can help in refining audience targeting and enhancing overall revenue.
